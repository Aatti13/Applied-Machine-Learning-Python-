# Logistic Regression

This is an algorithm mainly used for binary classification. It models probabiity of binary outcomes using multiple parameters or variables.
i. y belongs to {0,1}.

## Method-1

  A Mathematical approach to logistic regression.

In logistic regression we use the sigmoid function which is:

  ### y = 1/(1+exp(-&#952;<sup>T</sup>x))

For Logistic Regression, the cost function is the log-likelihood of the sigmoid function.

i.e: 

### f(&#952;<sup>T</sup>x) = (-1/m){ylog(h)+(1-y)log(1-h)}

We then perform gradient descent on this function to traverse the points.

### Code-1

```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as mplt
```

Now we are gonna define a sigmoid function

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

Now we make the Cost function

```python
def costFunction(X, y, theta):
  m = len(y)
  h = sigmoid(np.dot(X, theta))
  cost = (-1/m) * (np.dot(y, np.log(h)) + np.dot((1 - y), np.log(1 - h)))
  return cost
```

```python
def gradientDescent(X, y, theta, alpha, num_iterations):
    m = len(y)
    cost_history = []

    for i in range(num_iterations):
        h = sigmoid(np.dot(X, theta))
        gradient = np.dot(X.T, (h - y)) / m
        theta = theta - alpha * gradient
        cost = costFunction(X, y, theta)
        cost_history.append(cost)

    return theta, cost_history
```

```python
def plot_data_and_decision_boundary(X, y, theta):
    # Plot the data points
    mplt.scatter(X[y == 0][:, 1], X[y == 0][:, 2], c='red', label='Class 0')
    mplt.scatter(X[y == 1][:, 1], X[y == 1][:, 2], c='blue', label='Class 1')

    # Plot the decision boundary
    x_values = [np.min(X[:, 1]), np.max(X[:, 1])]
    y_values = - (theta[0] + np.dot(theta[1], x_values)) / theta[2]
    mplt.plot(x_values, y_values, label='Decision Boundary')

    mplt.xlabel('Feature 1')
    mplt.ylabel('Feature 2')
    mplt.legend()
    mplt.show()
```

```python
X = np.array([[1, 2, 1], [1, 3, 2], [1, 4, 3], [1, 5, 4]])
y = np.array([0, 0, 1, 1])

# Initialize parameters
theta = np.zeros(X.shape[1])
alpha = 0.01
num_iterations = 10000

# Perform gradient descent
theta, cost_history = gradientDescent(X, y, theta, alpha, num_iterations)

print("Optimized theta:", theta)
print("Cost history:", cost_history)

# Plot data and decision boundary
plot_data_and_decision_boundary(X, y, theta)
```
